```python
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt

file_path = r"C:\Users\LENOVO\Desktop\The\data\map.osm"  
graph = ox.graph_from_xml(file_path)

ox.plot_graph(graph)
```

```python
nodes = list(graph.nodes())
node1 = nodes[0]  
node2 = nodes[100] 

shortest_path = nx.shortest_path(graph, node1, node2, weight='length' , method="dijkstra")

ox.plot_graph_route(graph, shortest_path, route_linewidth=6, node_size=0, bgcolor='k')
```
